```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 16, 10
with open("optimized_signal_analysis2.csv", newline='') as file:
    result_list = list(csv.reader(file))

print(result_list)
```

```python
with open("generated_traffic_dataset_for_single_day.csv", newline='') as file:
    v_list = list(csv.reader(file))
print(len(v_list))
```

```python
print(v_list)
```

```python
with open("fixed_signal_analysis.csv", newline='') as file:
    result_list1 = list(csv.reader(file))
```

```python
print(len(result_list))
```

```python
optimized =[]
veh=[]
fixed=[]
tmp=[0,0]
tmpf=[0,0]
for i in range (1,len(result_list)):
    tmp[1]+=float(result_list[i][4])
    tmp[0]+=float(result_list[i][2])
    
    
    if i%3==0:
        optimized.append(tmp[0]/tmp[1])
        tmp[0]=0
        tmp[1]=0

for i in range (1,len(result_list1)):
    
    tmpf[0]+=float(result_list1[i][2])
    tmpf[1]+=float(result_list1[i][4])
    
    if i%3==0:
        fixed.append(tmpf[0]/tmpf[1])
        tmpf[0]=0
        tmpf[1]=0

for i in range(1,len(v_list)-1):
    veh.append(float(v_list[i][4])+float(v_list[i][1])+float(v_list[i][2])+float(v_list[i][3]))
```

```python
plt.figure()
plt.title("Analysis of Average delay")
plt.xlabel("Cycle")
plt.ylabel("Average delay of the vehicle in current cycle (s)")
plt.plot(range(len(fixed)),fixed,"-r",label="Fixed Signaling")
plt.plot(range(len(optimized)),optimized,"-g",label="Optimized Signaling")
plt.legend()


# plt.savefig('Avg_delay_analysis2.png')
```

```python
plt.figure()
plt.title("Total vehicles generated on all lanes")
plt.xlabel("Cycle")
plt.ylabel("Number of vehicles")
plt.plot(range(len(veh)),veh,"-b")
plt.legend()
plt.savefig('number_of_vehicles_analysis.png')
```
